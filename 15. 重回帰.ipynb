```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pystan
import seaborn as sns 
%matplotlib inline
plt.style.use("ggplot")
```

```python
df = pd.read_excel("./data/real_estate2.xlsx")
df.head()
```

```python
# 現在の年から築年数を求める
df["elapsed"] = 2018 - df["year"]
```

```python
# distanceの種類を求める
df["distance"].unique()
```

```python
# distanceの各値に対応する分数を得る
dis_arr = df["distance"].unique()
dis_dict = {dis_arr[0]: 10, dis_arr[1]: 15, dis_arr[2]: 5, dis_arr[3]: 20, dis_arr[4]: 30, dis_arr[5]: np.nan}
df["distance2"] = df["distance"].apply(lambda x: dis_dict[x])
df.head()
```

```python
# naを含むデータを取り除く
print(df.shape)
df = df.dropna()
print(df.shape)
```

```python
# 今回使う列のみを取得
df2 = df[["space", "elapsed", "distance2", "value"]]
df2.head()
```

多変量データの可視化を行うため，Seabornを利用する

```python
# 特徴量のペアグラフを作成
g = sns.PairGrid(df2)

# 下三角部分は，カーネル密度関数によるプロットを行う
# カーネル密度関数: 実データ点の周りに正規分布などを撒いてgridの各点の値を合算
g = g.map_lower(sns.kdeplot)

# 対角部分にはヒストグラムを描く
g = g.map_diag(sns.distplot, kde=False)

# 上三角部分には，散布図を描く
g = g.map_upper(plt.scatter)
```

縦軸がvalueの行に着目
- spaceが大きくなるほど若干物件価格も上がる
- 築年数が上がるほど物件価格がわかりやすく下がる
- 距離が長くなるほど物件価格が下がる

```python
stan_model = """
data {
    int N;
    real elapsed[N];
    real dis[N];
    real space[N];
    real Y[N];
}

parameters {
    real e;
    real d;
    real s;
    real b;
    real<lower=0> sigma;
}

model {
    real mu;
    for (n in 1:N) {
        mu = e * elapsed[n] + d * dis[n] + s * space[n] + b;
        Y[n] ~ normal(mu, sigma);
    }
}
"""
```

```python
sm = pystan.StanModel(model_code = stan_model)
```

```python
stan_data = {
    "N": df.shape[0], 
    "elapsed": df["elapsed"], 
    "dis": df["distance2"],
    "space": df["space"],
    "Y": df["value"]
}
```

```python
fit = sm.sampling(data = stan_data, iter=2000, warmup=500, chains=4, seed=123)
```

```python
fit
```

- e: 1年経過するごとに相場67万安くなる
- d: 1分遠くなるごとに相場57万安くなる
- s: 1平米増えるごとに24万安くなる

```python
fig = fit.plot()
```

いずれのパラメータもまんべんなくサンプリングしているので収束している
また，ピークがいずれも中央に来ていることから，データにも偏りが少ないことがわかる
