```python
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
p_a = 3.0 / 10.0 # aの袋を選んだときに赤を引く確率
p_b = 5.0 / 9.0   # bの袋を選んだ時に赤を引く確率
p_prior = 0.5 # どちらの袋をえらぶかの確率は同じ

# 0: blue, 1: red
data = [0,1,0,0,1,1,1]
```

```python
N_data = 2 # 幾つのデータから事後分布を求めるか

# aの袋を選んだ時の尤度関数
likelihood_a = bernoulli.pmf(data[:N_data], p_a) # pmf: 離散の確率密度関数

# bの袋を選んだ時の尤度関数
likelihood_b = bernoulli.pmf(data[:N_data], p_b)
```

```python
def post(N_data):

    # aの袋を選んだ時の尤度関数
    likelihood_a = bernoulli.pmf(data[:N_data], p_a) # pmf: 離散の確率密度関数


    # bの袋を選んだ時の尤度関数
    likelihood_b = bernoulli.pmf(data[:N_data], p_b)


    # 事前分布をセット
    pa_posterior = p_prior
    pb_posterior = p_prior


    # 得られたデータによる尤度関数の積をとる
    pa_posterior *= np.prod(likelihood_a) 
    pb_posterior *= np.prod(likelihood_b)


    # 規格化定数
    norm = pa_posterior + pb_posterior


    # 事後分布をグラフで表示
    df = pd.DataFrame([pa_posterior / norm, pb_posterior / norm], columns=['post'])
    x = np.arange(df.shape[0])
    
    return df, x

print("aの袋を選んだ時に赤を引く確率:", round(p_a, 2))
print("bの袋を選んだ時に赤を引く確率:", round(p_b, 2))
print("0: blue, 1: red")
print(data)

fig, ax = plt.subplots(1, 7, figsize=(30, 5))
for i in range(len(data)):
    df, x = post(i+1)
    ax[i].bar(x, df["post"])
    ax[i].set_xticks(x)
    ax[i].set_xticklabels(["a", "b"])
    ax[i].set_title(f"n={i+1}")

plt.show()
```
