```python
import numpy as np
from scipy import stats
from scipy import optimize as opt
from scipy.stats import beta, uniform
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
np.random.seed(123)
```

Beta(a, b)分布を描く

```python
a, b = 1.5, 2.0

# percent point function:与えた値をとるx値を返す
x = np.linspace(beta.ppf(0.001, a, b), beta.ppf(0.999, a, b), 100)
plt.plot(x, beta.pdf(x, a, b))
```

Beta分布の最高点を求める

```python
f = beta(a=a, b=b).pdf

# マイナスをつけて最小点を求める
res = opt.fmin(lambda x: -f(x), 0.3)

# 分布の頂点
y_max = f(res)
print(y_max)
```

```python
NMCS = 5000  # ステップ数
x_mcs = uniform.rvs(size=NMCS) # 5000個のサンプリング
r = uniform.rvs(size=NMCS) * y_max # 0~y_maxまでの値のサンプリング
accept = x_mcs[r <= f(x_mcs)]
plt.hist(accept, normed=True, bins=30, rwidth=0.8, label="rejection sampling")

# Beta分布も描画する
x = np.linspace(beta.ppf(0.001, a, b), beta.ppf(0.999, a, b), 100)
plt.plot(x, beta.pdf(x, a, b), label="Target dist")
plt.legend()
```
