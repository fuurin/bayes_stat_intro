```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pystan
from scipy.stats import mstats
%matplotlib inline
plt.style.use("ggplot")
```

```python
df = pd.read_excel("./data/multilevel_modeling.xlsx")
df.head()
```

```python
print(df["id"].unique())
groups = df.groupby(df["id"])
for name, group in groups:
    plt.plot(group["age"], group["height"], label=name)
plt.legend()
```

人によって身長の伸び方の個人差があることがわかる

```python
stan_model = """
data {
    int N;
    int N_id;
    real X[N];
    real Y[N];
    
    // 生徒のIDごとに個人差の分散を求めるため，idリストを受け取る
    int<lower=1, upper=N_id> s_id[N];
}

parameters {
    real a0;
    real b0;
    real a_id[N_id];
    real b_id[N_id];
    real<lower=0> s_a;
    real<lower=0> s_b;
    real<lower=0> s_Y;
}

// 個人ごとの傾きを作るためのブロック
transformed parameters {
    real a[N_id];
    real b[N_id];
    for (n in 1:N_id) {
        a[n] = a0 + a_id[n];
        b[n] = b0 + b_id[n];
    }
}

model {
    for (id in 1:N_id) {
        a_id[id] ~ normal(0, s_a);
        b_id[id] ~ normal(0, s_b);
    }
    
    for (n in 1:N) {
        Y[n] ~ normal(a[s_id[n]] * X[n] + b[s_id[n]], s_Y);
    }
}
"""
```

```python
sm = pystan.StanModel(model_code = stan_model)
```

```python
stan_data = {
    "N": df.shape[0], 
    "N_id": 15,
    "X": df["age"], 
    "Y": df["height"],
    "s_id": df["id"]
}
```

```python
fit = sm.sampling(data = stan_data, iter=3000, warmup=1000, chains=3, seed=123)
```

```python
fit
```

a0から伸び盛りの頃の平均の身長の伸びは1年に5.77cmほどだということがわかる  
a_id[i]は, i番目の人の平均と比較した伸びやすさを意味する

```python
fig = fit.plot()
```

```python
ms_a = fit.extract("a")["a"]
ms_b = fit.extract("b")["b"]
```

```python
x = np.arange(18)
df_b = pd.DataFrame([])
for i in range(18):
    df_b[i] = ms_a[:, 0] * x[i] + ms_b[:, 0]
```

```python
low_y50, high_y50 = mstats.mquantiles(df_b, [0.25, 0.75], axis=0)
low_y95, high_y95 = mstats.mquantiles(df_b, [0.025, 0.975], axis=0)
```

```python
df_0 = groups.get_group(1)
df_0.head()
```

```python
plt.plot(df_0["age"], df_0["height"])
plt.fill_between(x, low_y50, high_y50, alpha=0.6, color="darkgray")
plt.fill_between(x, low_y95, high_y95, alpha=0.3, color="gray")
```
