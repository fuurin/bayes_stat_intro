```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pystan
from scipy.stats import mstats
%matplotlib inline
plt.style.use("ggplot")
```

```python
df = pd.read_excel("./data/temperature_series.xlsx")
df.head()
```

```python
plt.plot(df["x"], df["y"])
```

人によって身長の伸び方の個人差があることがわかる

```python
stan_model = """
data {
    int T;             // 時系列データなのでNでなくT
    int T_new;   // 何年先を予測するか．来年のを知りたいのであとで1を入れる
    real Y[T];     // 観測データ
}

parameters {
    real mu[T];                   // 直接の期待値
    real<lower=0> s_mu; // 変化の具合
    real<lower=0> s_Y;    // 測定誤差
}

model {
    for (t in 2:T) {
        mu[t] ~ normal(mu[t-1], s_mu);
    }
    
    for (t in 1:T) {
        Y[t] ~ normal(mu[t], s_Y);
    }
}

generated quantities {
    real mu_new;
    real Y_new[T+T_new];
    for (t in 1:T) {
        Y_new[t] = normal_rng(mu[t], s_Y);
    }
    mu_new = normal_rng(mu[T], s_mu);
    Y_new[T+T_new] = normal_rng(mu_new, s_Y);
}
"""
```

```python
sm = pystan.StanModel(model_code = stan_model)
```

```python
stan_data = {"T": df.shape[0], "T_new": 1, "Y": df["y"]}
```

```python
fit = sm.sampling(data = stan_data, iter=3000, warmup=1500, chains=3, seed=123)
```

```python
print(fit)
```

a0から伸び盛りの頃の平均の身長の伸びは1年に5.77cmほどだということがわかる  
a_id[i]は, i番目の人の平均と比較した伸びやすさを意味する

```python
fig = fit.plot()
```

```python
Y_new_arr = fit.extract("Y_new")["Y_new"]
```

```python
low_y50, high_y50 = mstats.mquantiles(Y_new_arr, [0.25, 0.75], axis=0)
low_y95, high_y95 = mstats.mquantiles(Y_new_arr, [0.025, 0.975], axis=0)
```

```python
plt.plot(df["x"], df["y"])
x = df["x"].values            # 年
x = np.append(x, 2017) # 来年
plt.fill_between(x, low_y50, high_y50, alpha=0.6, color="darkgray")
plt.fill_between(x, low_y95, high_y95, alpha=0.3, color="gray")
```
